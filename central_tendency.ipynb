```python
#importing packages
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
```

```python
#reading housing data (rent) with pandas
brooklyn_one_bed = pd.read_csv('brooklyn-one-bed.csv')
brooklyn_price = brooklyn_one_bed['rent']

manhattan_one_bed = pd.read_csv('manhattan-one-bed.csv')
manhattan_price = manhattan_one_bed['rent']

queens_one_bed = pd.read_csv('queens-one-bed.csv')
queens_price = queens_one_bed['rent']
```

```python
# Calculating mean
brooklyn_mean = np.mean(brooklyn_price)
manhattan_mean = np.mean(manhattan_price)
queens_mean = np.mean(queens_price)

# Displaying average rent price of apartments in different cities
print(f'The average rent price in Brooklyn is ${brooklyn_mean}\n')
print(f'The average rent price in Manhattan is ${manhattan_mean}\n')
print(f'The average rent price in Queens is ${queens_mean}\n')
```

```python
# Calculating median
brooklyn_median = np.median(brooklyn_price)
manhattan_median = np.median(manhattan_price)
queens_median = np.median(queens_price)

# Displaying median/mid-point rent price of apartments in different cities
print(f'The median price in Brooklyn is ${brooklyn_median}\n')
print(f'The median price in Manhattan is ${manhattan_median}\n')
print(f'The median price in Queens is ${queens_median}\n')
```

```python
# Calculating mode
brooklyn_mode = stats.mode(brooklyn_price)
manhattan_mode = stats.mode(manhattan_price)
queens_mode = stats.mode(queens_price)

# Getting brooklyn_mode value and it's count
brooklyn_mode_value = brooklyn_mode[0][0]
brooklyn_mode_count = brooklyn_mode[1][0]
print(f'The mode or the most common rent price and its frequency of Brooklyn\'s apartments are: ${brooklyn_mode_value} and {brooklyn_mode_count} respectively\n')
# Getting manhattan_mode value and it's count
manhattan_mode_value = manhattan_mode[0][0]
manhattan_mode_count = manhattan_mode[1][0]
print(f'The mode or the most common rent price and its frequency of Manhattan\'s apartments are: ${manhattan_mode_value} and {manhattan_mode_count} respectively\n')
# Getting queens_mode value and it's count
queens_mode_value = queens_mode[0][0]
queens_mode_count = queens_mode[1][0]
print(f'The mode or the most common rent price and its frequency of Queens\'s apartments are: ${queens_mode_value} and {queens_mode_count} respectively\n')

```

```python
# Displaying central tendency - mean/average, median and mode in histogram of Brooklyn city
plt.hist(brooklyn_price,range=(0,10000), bins=22,  edgecolor='black')
plt.title("Brooklyn apartment rental chart")
plt.xlabel("Rent price")
plt.ylabel("Count")
plt.axvline(brooklyn_mean, color='r', linestyle='solid', linewidth=3, label="Mean")
plt.axvline(brooklyn_median, color='y', linestyle='dotted', linewidth=3, label="Median")
plt.axvline(brooklyn_mode_value, color='orange', linestyle='dashed', linewidth=3, label="Mode")
plt.legend()

plt.show()
```

```python
# Displaying central tendency - mean/average, median and mode in histogram of Manhattan city
plt.hist(manhattan_price,range=(0,10000), bins=22,  edgecolor='black')
plt.title("Manhattan apartment rental chart")
plt.xlabel("Rent price")
plt.ylabel("Count")
plt.axvline(manhattan_mean, color='r', linestyle='solid', linewidth=3, label="Mean")
plt.axvline(manhattan_median, color='y', linestyle='dotted', linewidth=3, label="Median")
plt.axvline(manhattan_mode_value, color='orange', linestyle='dashed', linewidth=3, label="Mode")
plt.legend()

plt.show()
```

```python
# Displaying central tendency - mean/average, median and mode in histogram of Queens city
plt.hist(manhattan_price,range=(0,10000), bins=7,  edgecolor='black')
plt.title("Queens apartment rental chart")
plt.xlabel("Rent price")
plt.ylabel("Count")
plt.axvline(queens_mean, color='r', linestyle='solid', linewidth=3, label="Mean")
plt.axvline(queens_median, color='y', linestyle='dotted', linewidth=3, label="Median")
plt.axvline(queens_mode_value, color='orange', linestyle='dashed', linewidth=3, label="Mode")
plt.legend()

plt.show()
```
